```python
#Collecting the HOST details
hostData = []
for i in coll.find({},{"_id":1, "host":1}):
  hostData.append(i)
```

```python
dfHost = pd.DataFrame(hostData)
```

```python
hostColumns= {'_id':[],'host_id':[], 'host_url':[], 'host_name':[], 'host_location':[],"host_response_time":[],
               'host_thumbnail_url':[], 'host_picture_url':[], 'host_neighbourhood':[], 'host_response_rate':[],
               'host_is_superhost':[], 'host_has_profile_pic':[], 'host_identity_verified':[], 'host_listings_count':[],
               'host_total_listings_count':[], 'host_verifications':[]}

for i in dfHost["_id"]:
    hostColumns["_id"].append(i)
for i in dfHost["host"]:
    hostColumns["host_id"].append(i["host_id"])
    hostColumns["host_url"].append(i["host_url"])
    hostColumns["host_name"].append(i["host_name"])
    hostColumns["host_location"].append(i["host_location"])
    hostColumns["host_response_time"].append(i.get("host_response_time"))
    hostColumns["host_thumbnail_url"].append(i["host_thumbnail_url"])
    hostColumns["host_picture_url"].append(i["host_picture_url"])
    hostColumns["host_neighbourhood"].append(i["host_neighbourhood"])
    hostColumns["host_response_rate"].append(i.get("host_response_rate"))
    hostColumns["host_is_superhost"].append(i["host_is_superhost"])
    hostColumns["host_has_profile_pic"].append(i["host_has_profile_pic"])
    hostColumns["host_identity_verified"].append(i["host_identity_verified"])
    hostColumns["host_listings_count"].append(i["host_listings_count"])
    hostColumns["host_total_listings_count"].append(i["host_total_listings_count"])
    hostColumns["host_verifications"].append(i["host_verifications"])
dfHostUpdate = pd.DataFrame(hostColumns)
```

```python
#Collecting the Address
address = []
for i in coll.find({},{"_id":1, "address":1}):
  address.append(i)
```

```python
dfAddress = pd.DataFrame(address)
```

```python
addressColumn = {"_id": [], "street": [], "suburb": [], 'government_area': [], "market": [],
                 "country": [], "country_code": [], "location_type": [],
                 "longitude": [], "latitude": [], 'is_location_exact': []}
for i in dfAddress["_id"]:
  addressColumn["_id"].append(i)

for j in dfAddress["address"]:
  addressColumn["street"].append(j["street"])
  addressColumn['suburb'].append(j['suburb'])
  addressColumn["government_area"].append(j["government_area"])
  addressColumn["market"].append(j["market"])
  addressColumn["country"].append(j["country"])
  addressColumn["country_code"].append(j["country_code"])
  addressColumn["location_type"].append(j["location"]["type"])
  addressColumn["longitude"].append(j["location"]["coordinates"][0])
  addressColumn["latitude"].append(j["location"]["coordinates"][1])
  addressColumn["is_location_exact"].append(j["location"]["is_location_exact"])
```

```python
dfAddressUpdate = pd.DataFrame(addressColumn)
```

```python
# Checking the empty values ('') for all features
id_e=[]
stre_e=[]
suburb_e=[]
gov_e=[]
mark_e=[]
cntry_e=[]
cntry_co_e=[]
loc_ty_e=[]
long_e=[]
lat_e=[]
is_loc_tr_e=[]
for index,row in dfAddressUpdate.iterrows():
    if row["_id"] == '':
        id_e.append(index)

    if row["street"] == '':
        stre_e.append(index)

    if row["suburb"] == '':
        suburb_e.append(index)

    if row["government_area"] == '':
        gov_e.append(index)

    if row["market"] == '':
        mark_e.append(index)

    if row["country"] == '':
        cntry_e.append(index)

    if row["country_code"] == '':
        cntry_co_e.append(index)

    if row["location_type"] == '':
        loc_ty_e.append(index)

    if row["longitude"] == '':
        long_e.append(index)

    if row["latitude"] == '':
        lat_e.append(index)

    if row["is_location_exact"] == '':
        is_loc_tr_e.append(index)

empty_columns=[id_e,stre_e,suburb_e,gov_e,mark_e,cntry_e,cntry_co_e,loc_ty_e,long_e,lat_e,is_loc_tr_e]
for i in empty_columns:
    print(len(i))
```

```python
# Handling the empty values

dfAddressUpdate["suburb"] = dfAddressUpdate["suburb"].replace({'':"Not Specified"})
dfAddressUpdate["market"] = dfAddressUpdate["market"].replace({'':"Not Specified"})
dfAddressUpdate["is_location_exact"] = dfAddressUpdate["is_location_exact"].map({False:"No", True:"Yes"})
```

```python
#Collecting the availability
available = []

for i in coll.find({}, {"_id": 1, "availability": 1}):
  available.append(i)

dfAvailable = pd.DataFrame(available)
```

```python
availableColumn = {"_id": [], 'availability_30': [], 'availability_60': [], 'availability_90': [], 'availability_365': []}

for i in dfAvailable["_id"]:
  availableColumn["_id"].append(i)

for j in dfAvailable["availability"]:
  availableColumn["availability_30"].append(j['availability_30'])
  availableColumn["availability_60"].append(j['availability_60'])
  availableColumn["availability_90"].append(j['availability_90'])
  availableColumn["availability_365"].append(j['availability_365'])

dfAvailableUpdate = pd.DataFrame(availableColumn)
```

```python
#Collecting the Amenities

amenities = []

for i in coll.find({},{"_id":1, 'amenities': 1}):
  amenities.append(i)

dfAmenities = pd.DataFrame(amenities)
```

```python
def sortAmenities(x):
  a = x
  a.sort()
  return a

dfAmenities["amenities"]= dfAmenities["amenities"].apply(lambda x: sortAmenities(x))
```

```python
dfAmenities.isna().sum()
```

```python
#Merging all the dataframes
#df
#dfHostUpdate
#dfAddressUpdate
#dfAvailableUpdate
#dfAmenities

dfAirbnb = pd.merge(df, dfHostUpdate, on = "_id")
dfAirbnb = pd.merge(dfAirbnb, dfAddressUpdate, on = "_id")
dfAirbnb = pd.merge(dfAirbnb, dfAvailableUpdate, on = "_id")
dfAirbnb = pd.merge(dfAirbnb, dfAmenities, on = "_id")
```
